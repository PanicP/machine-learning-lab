1 Reduction
Take the array ‘traind’ and ...
a) Compute the pixel argmax for pixel 0,0 over all samples
b) Compute the row-wise mean over all samples
c) Compute the column-wise sum over all samples

```python
import matplotlib.pyplot as plt
import numpy as np

traind, testd, trainl, testl = np.load(open("mnist.npz", "rb")).values()

#a
a = np.argmax(traind[:,0:1,0:1])
#b
b = np.mean(traind, axis=2)
#c
c = np.sum(traind, axis=1)
print('a', a)
print('b', b)
print('c', c)
```

2 Matplotlib
a) plot the function 1/x between 1 and 5 using 100 support points!
b) generate a scatter plot of the same data as in a)!
c) generate a bar plot of the same data as in a)!
d) plot 1/x and √x together in a single scatter plot, same range as before

```python
import matplotlib.pyplot as plt
import numpy as np

traind, testd, trainl, testl = np.load(open("mnist.npz", "rb")).values()
# arr = np.arange(1,6,1)
(_,ax) = plt.subplots(2, 2)
arr = np.linspace(1,5,100)
# print(arr)
#a
ax[0, 0].set_title("a")
ax[0, 0].plot(arr, 1/arr)

#b
ax[0, 1].set_title("b")
ax[0, 1].scatter(arr, 1/arr)

#c
ax[1, 0].set_title("c")
ax[1, 0].bar(arr, 1/arr)

#c
ax[1, 1].set_title("d")
ax[1, 1].scatter(arr, 1/arr)
ax[1, 1].scatter(arr, np.sqrt(arr))


plt.show()
```
