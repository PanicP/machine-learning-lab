```python
import numpy as np
import matplotlib.pyplot as plt

f, ax = plt.subplots(1,1)
data = np.linspace(1,5,100)
ax.plot(data, 1/data)
plt.show()

f, ax = plt.subplots(1,1)
data = np.linspace(1,5,100)
ax.scatter(data, 1/data)
plt.show()

f, ax = plt.subplots(1,1)
data = np.linspace(1,5,100)
ax.bar(data, 1/data)
plt.show()

f, ax = plt.subplots(1,1)
data = np.linspace(1,5,100)
ax.plot(data, 1/data)
ax.plot(data, np.sqrt(data))
plt.show()

f, ax = plt.subplots(1,1)
data = np.random.uniform(0,1,100)
ax.hist(data)
plt.show()
```
